```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url="https://en.wikipedia.org/wiki/Python_(programming_language)"
page= requests.get(url) #extraemos el contenido de la pagina
soup= BeautifulSoup(page.content , "html.parser")

head= soup.find_all(["h1","h2","h3","h4","h5","h6"])
print(head)
lista=[]
for i in head:
    lista.append(i.text)
encabezado = pd.DataFrame({"encabezados":lista})
print(len(encabezado))   

head= soup.find_all("h1") #buscamos todos los encabezados h1 
h1=[]
for i in head:
    h1.append(i.text)
        
head= soup.find_all("h2") #buscamos todos los encabezados h2
h2=[]
for i in head:
    h2.append(i.text)
    
head= soup.find_all("h3") #buscamos todos los encabezados h3
h3=[]
for i in head:
    h3.append(i.text)    
# igualando la longitud de datos para cada lista de encabezados
while len(h1)<len(h3) : 
    h1.append("-")
    if len(h2)<len(h3):
          h2.append("-")
#eliminando los "[edit]" y  "\n" de la lista .        
for i in range(len(h3)):
    if "[edit]" in h3[i]:
        h3[i]=h3[i][:-6]
    if "\n" in h3[i]:
        h3[i]=h3[i][1:-1]
    if "[edit]" in h2[i]:
        h2[i]=h2[i][:-6]

diccionario = {'h1' : h1, 'h2' : h2, 'h3': h3 } #creamos un diccionario con todos los encabezados
dic = pd.DataFrame(diccionario,columns=["h1","h2","h3"]) #creamos un diccionario
dic.to_csv('encabezados.csv') #convertimos en csv 

#leyendo la data creada de todos los encabezados 
ENCABEZADOS = pd.read_csv('encabezados.csv',skiprows=1,            
                 names=['UID','h1', 'h2', 'h3'] , index_col='UID')
ENCABEZADOS
```

```python
#-Descargar la imagen de Guido van Rossum  y 
# almacenarlo en un archivo “guido.jpg”
from bs4 import BeautifulSoup
import requests
import urllib.request, urllib.parse, urllib.error


url="https://en.wikipedia.org/wiki/Python_(programming_language)"
page= requests.get(url) 
soup= BeautifulSoup(page.content , "html.parser")

#buscando el enlace de la imagen en la pagina web
conten = soup.find_all('div' , class_="thumbinner") 

lista = []
for t in conten : 
    sub = t.find('img' , class_="thumbimage").get("src")
    lista.append(sub)
    #print(sub) 
print(lista)
#enlace se encuentra en el primer elemento
#descargando la imagen
img = urllib.request.urlopen('https:'+lista[0]).read()#abriendo la url y descargando el contenido.
f = open('g.jpg', 'wb')
f.write(img)
f.close() 
```
